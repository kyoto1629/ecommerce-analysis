```python
# 8. Анализ цен товаров

# Исследуем распределение цен:
# - минимальные и максимальные значения;
# - среднюю и медианную цену;
# - наличие выбросов;
# - цены по типам событий.
```

```python
df['price'].describe()
```

```python
# 8.1 Средняя и медианная цена
```

```python
print('Средняя цена:', round(df['price'].mean(), 2))
print('Медианная цена:', round(df['price'].median(), 2))
```

```python
# Основные наблюдения:

# Мы видим классический случай правосторонней асимметрии (positive skew). Средняя цена (291.82) значительно выше медианной (164.98) — разрыв составляет около 77%.

# Это означает, что распределение цен не является нормальным.
```

```python
# 8.2 Гистограмма распределения цен

# Так как в e-commerce часто есть выбросы, сначала посмотрим на распределение в целом.
```

```python
sns.histplot(df['price'], bins=50)
plt.title('Распределение цен товаров')
plt.xlabel('Цена')
plt.ylabel('Количество')
plt.show()
```

```python
# Основные наблюдения:

# График подтверждает гипотезу о сильной правосторонней асимметрии, о которой мы говорили на прошлом шаге. Визуально это классическая «L-образная» кривая с длинным «хвостом».

# Доминирование дешевого сегмента:
## Огромный первый столбец (цена ~50-100) достигает отметки в 400_000+ событий. Это значит, что львиная доля всего трафика приходится на самые дешевые товары.

# Резкое падение интереса:
## Уже в диапазоне 200–300 количество событий падает в несколько раз. К отметке 500 активность снижается до уровня ~50_000–80_000 событий. Это подтверждает, что медиана (~165) — это реальная граница «основного потока».

# Одинокий выброс:
## В районе 2500 виден маленький, но отдельный столбик. Вероятно, это единичные товары эксклюзивной категории или ошибка данных (например, цена за опт вместо розницы), которые резко выбиваются из общего ряда.

# Бизнес-модель площадки явно опирается на массовый сегмент (до 200 единиц валюты). Попытки продавать товары дороже 1000 встречают очень слабый отклик (с точки зрения количества просмотров/кликов).
```

```python
# 8.3 Анализ цен без сильных выбросов

# Ограничим цены 99-м перцентилем, чтобы лучше увидеть основную массу товаров.
```

```python
price_99 = df['price'].quantile(0.99)
print('99-й перцентиль цены:', round(price_99, 2))
```

```python
df_price_filtered = df[df['price'] <= price_99]

sns.histplot(df_price_filtered['price'], bins=50)
plt.title('Распределение цен (до 99-го перцентиля)')
plt.xlabel('Цена')
plt.ylabel('Количество')
plt.show()
```

```python
# Основные наблюдения:

# Четкая структура основной массы
## Отсечение топ-1% самых дорогих товаров позволило четко увидеть основную картину. График по-прежнему демонстрирует правостороннюю асимметрию, но теперь хорошо видны внутренние закономерности, которые ранее размывались из-за экстремальных выбросов.

# Ядро спроса
## Максимальная концентрация событий (~300 тыс.) приходится на диапазон 100–150. Это абсолютное ядро ассортимента, генерирующее основную долю трафика и взаимодействий.

# Вторичные пики и «плато»
## На графике отчетливо видны локальные повышения активности в районах ~750 и ~950–1000. После плавного спада от 250 до 500, распределение переходит в более равномерный, «плоский» участок вплоть до ~1250, после чего снова резко снижается к верхней границе фильтра.

# Психологическое ценообразование
## Вторичные пики вокруг 750 и 1000 с высокой вероятностью отражают использование маркетологами ценовых барьеров: 799, 999, 1499 и т.д. Пользователи активнее кликают и добавляют в корзину товары, укладывающиеся в эти круглые психологические пороги.
```

```python
# 8.4 Boxplot цен
```

```python
sns.boxplot(x=df_price_filtered['price'])
plt.title('Boxplot цен товаров')
plt.xlabel('Цена')
plt.show()
```

```python
# Основные наблюдения:

# Визуализация "длинного хвоста"
## Самый заметный элемент графика — это густая черная линия справа, тянущаяся от отметки ~750 до ~1700 - так отображаются выбросы. Линия выглядит как толстый сплошной маркер, это говорит о том, что "дорогих" наблюдений очень много. Они плотно накладываются друг на друга, подтверждая, что в данном датасете есть огромный пласт товаров с ценой от 750 до 1700, которые статистически считаются аномалиями относительно основной массы.

# IQR
## Синий прямоугольник (интерквартильный размах) охватывает диапазон примерно от 50 до 350.
## Это означает, что 50% всех событий в датасете происходит с товарами именно в этом ценовом коридоре. Это еще раз подтверждает, что основной объем работы сайта — это обработка заявок на товары стоимостью до 350 единиц.

# Позиция медианы
## Темная вертикальная линия внутри синего прямоугольника (медиана) находится левее центра (около 165). Это классический признак правосторонней асимметрии. "Нормальные" товары сгруппированы в начале координат, но их распределение неравномерно — дешевых товаров значительно больше, чем тех, что стоят ближе к 350.

# График наглядно показывает, что у нас как бы два кластера поведения.
## Кластер 1 (Синий ящик): Массовый поток, дешевые товары, высокая частота.
## Кластер 2 (Черная линия): Нишевые, дорогие товары. Их меньше по количеству событий, но они есть, и их много (иначе линия была бы пунктирной).
```

```python
# 8.5 Сравнение цен по типам событий
```

```python
df.groupby('event_type')['price'].agg(['count', 'mean', 'median', 'min', 'max']).round(2)
```

```python
sns.boxplot(data=df[df['price'] <= price_99], x='event_type', y='price')
plt.title('Распределение цен по типам событий')
plt.xlabel('Тип события')
plt.ylabel('Цена')
plt.show()
```

```python
# Основные наблюдения:

# Рост "качества" чека по воронке
## Сравнивая статистику, мы видим интересный тренд: Median(view) < Median(cart) = Median(purchase).
## Медиана просмотров: ~165
## Медиана добавлений в корзину и покупок: ~175
## Это означает, что пользователи с большей вероятностью добавляют в корзину и покупают товары из более высокого ценового сегмента (ближе к 175+), чем просто просматривают их. Самые дешевые товары (диапазон 0–150) часто просматриваются, но реже попадают в корзину относительно своей доли в общем трафике.

# Идентичность корзины и покупки
## Медианные значения для cart и purchase совпадают, а средние значения очень близки.
## Это говорит о том, что цена товара не является барьером на этапе чекаута. Если пользователь добавил товар в корзину, вероятность его покупки практически не зависит от цены (в рамках этого распределения).

# Одинаковый "потолок" цен
## Максимальная цена и форма "хвоста" выбросов одинаковы для всех трех типов событий.
## Пользователи готовы не только смотреть, но и покупать дорогие товары. Наличие дорогих покупок подтверждает, что на сайте есть платежеспособная аудитория, просто таких транзакций меньше в абсолютном количестве.

# Фильтрация "дешевого мусора"
## Раз медиана просмотров ниже медианы покупок, это может указывать на то, что в категории самых дешевых товаров (0–100) много "информационного шума" (аксессуары, мелкие детали), которые пользователи смотрят, но не покупают как основной товар. Либо же пользователи сравнивают несколько дешевых вариантов, но в корзину кладут один, более качественный (и дорогой).
```

```python
# 10.3 Средняя цена по брендам

# Оставим только бренды, у которых достаточно наблюдений, чтобы анализ был более устойчивым.
```

```python
brand_stats = (
    df[df['brand'] != 'unknown']
    .groupby('brand')
    .agg(
        events=('brand', 'count'),
        avg_price=('price', 'mean')
    )
    .query('events >= 1000')
    .sort_values('avg_price', ascending=False)
    .round(2)
)

brand_stats.head(15)
```

```python
top_expensive_brands = brand_stats.head(15)

sns.barplot(x=top_expensive_brands['avg_price'], y=top_expensive_brands.index)
plt.title('Топ-15 брендов по средней цене')
plt.xlabel('Средняя цена')
plt.ylabel('Бренд')
plt.show()
```

```python
# Основные наблюдения:

# Четкое ценовое расслоение брендов
## Средние цены варьируются от ~620 (acer) до ~1640 (peda). Это почти трехкратная разница внутри топ-15, что говорит о сильном разделении брендов по позиционированию на площадке.

# Лидеры — нишевые или узкоспециализированные марки
## peda (~1640) и omabelle (~1522) возглавляют рейтинг, но при этом имеют относительно небольшую аудиторию (~1.1k и ~2.7k событий). Скорее всего, это премиальные нишевые бренды. Их высокая средняя цена не «размывается» дешевыми товарами.

# Apple - лидер трафика, но не лидер по средней цене
## Несмотря на абсолютное доминирование в просмотрах и покупках, Apple занимает лишь 6-е место (~829). Это классический эффект диверсификации: у бренда огромный диапазон цен — от недорогих чехлов и кабелей до MacBook Pro. Большое количество дешевых взаимодействий тянет среднее значение вниз.

# Умная фильтрация данных
## Условие events >= 1000 в коде критически важно. Оно отсекает бренды с 1–2 дорогими товарами, которые могли бы искусственно завысить рейтинг. Мы видим только те марки, которые имеют устойчивое присутствие на площадке.

# Разные бизнес-модели брендов
## peda/omabelle/dji работают по модели «мало показов - высокий чек». Их аудитория целевая, приходит за конкретным дорогим продуктом.
## apple/acer/canon работают по модели «массовый охват - средний чек». Они генерируют основной трафик, но зарабатывают на объеме и экосистеме.
```

```python
# 11. Анализ пользователей

# Теперь посмотрим на поведение пользователей:
# - сколько действий совершает один пользователь;
# - насколько равномерно распределена активность.
```

```python
# 11.1 Количество событий на пользователя
```

```python
user_activity = df.groupby('user_id').size().reset_index(name='event_count')
user_activity.head()
```

```python
user_activity['event_count'].describe()
```

```python
# Основные наблюдения:

# Медиана = 1 — это главный инсайт
## Половина всех пользователей (50%) совершила ровно одно действие за весь период наблюдений. Это означает глубокого вовлечения у массовой аудитории нет.

# Среднее (1.89) выше медианы (1.00)
## Классический признак правосторонней асимметрии. Небольшая группа очень активных пользователей «тянет» среднее значение вверх. Если бы распределение было равномерным, среднее и медиана были бы близки.

# 75% пользователей — это 1–2 события
## Третий квартиль (75%) показывает, что только каждый четвертый пользователь делает больше двух действий. Остальные 75% — это «поверхностный» трафик с минимальной глубиной взаимодействия.

# Максимум = 478 событий
## Самый активный пользователь совершил почти 500 действий. Это может быть:
## Профессиональный закупщик или реселлер
## Корпоративный клиент с общим аккаунтом
## Бот или парсер контента
## Реально очень вовлеченный пользователь.

#Высокое стандартное отклонение (2.09)
## При среднем 1.89 отклонение 2.09 говорит о том, что данные крайне неоднородны. Большинство значений сгруппировано около 1–2, но есть «хвост» из пользователей с 50, 100, 200+ событиями, которые и создают такой разброс.

# Проблема удержания (retention)
## Если 50% пользователей делают одно действие и уходят, возможно, сайт не цепляет с первого раза.
```

```python
# 11.2 Распределение активности пользователей

# Так как в данных могут быть очень активные пользователи, ограничим график 99-м перцентилем.
```

```python
user_99 = user_activity['event_count'].quantile(0.99)
print('99-й перцентиль событий на пользователя:', user_99)
```

```python
sns.histplot(user_activity[user_activity['event_count'] <= user_99]['event_count'], bins=50)
plt.title('Распределение количества событий на пользователя')
plt.xlabel('Количество событий')
plt.ylabel('Количество пользователей')
plt.show()
```

```python
# Основные наблюдения:

# Эффект «Степенного закона» (Power Law)
## График демонстрирует классическое экспоненциальное затухание.
## Первый столбец (1 событие): Доминирует абсолютно. Здесь скопилось ~720 000 пользователей. Если вспомнить, что всего у нас ~1.16 млн пользователей, это означает, что ~62% аудитории совершает всего одно действие и исчезает.
## Резкий спад: Уже на втором шаге (2 события) количество пользователей падает более чем в 3 раза (~220 тыс.).
## Дальнейшее падение: К 3 событиям остается ~90 тыс. пользователей, к 4 — ~50 тыс. График быстро «прижимается» к оси X.

# Поведение «массового» пользователя
## Для 99% аудитории (мы отсекли самый активный 1%) характерна поверхностная активность. Большинство не погружается в каталог. Они заходят, делают 1–2 клика и уходят. Это типичная картина для e-commerce с высоким процентом «случайного» или рекламного трафика.

# Существование «китов» (очень активных пользователей)
## Тот факт, что мы вынуждены были обрезать график на 10 событиях (99-й перцентиль), а ранее мы видели максимум в 478 событий, говорит о том, что есть узкая прослойка сверхактивных пользователей. Они создают «длинный хвост» распределения. Именно на них, скорее всего, держится значительная часть повторных продаж.

# Проблема «Одного касания» (one-and-done)
## Поскольку ~60% пользователей имеют только 1 событие, и мы знаем из первого графика, что 95% всех событий — это view (просмотр), можно сделать вывод:
## Большинство людей заходят на карточку товара, смотрят и уходят, не совершая даже второго действия.
```

```python
# Основные наблюдения:
## Тотальное доминирование «однокликовых» сессий
## Это самый показательный график на данный момент. Ключевая метрика: 25%, 50% и 75% перцентили равны 1.00.
## Это означает, что минимум 75% всех сессий состоят из ровно одного события. Пользователь заходит на сайт, совершает одно действие (почти всегда view) и сессия завершается.

# Среднее (1.13) практически равно медиане (1.00)
## В отличие от анализа на уровне пользователей, здесь разрыв минимален. Стандартное отклонение всего 0.44 — это очень низкий показатель, который подтверждает: данные крайне однородны.

# Максимум = 81 событие в сессии
## Даже самая длинная сессия содержит «всего» 81 событие. Для сравнения: на уровне пользователя мы видели максимум 478 событий. Это важный сигнал: даже очень активные пользователи не «сидят» на сайте часами в одной сессии. Они возвращаются многократно, но каждый отдельный визит относительно короток.

# Масштаб явления
## Всего в датасете ~1.93 млн сессий. Если 75% из них — это одно событие, то мы говорим о ~1.45 млн «холостых» визитов, которые не привели ни к продолжению просмотра, ни к добавлению в корзину.

# Экстремально высокий Bounce Rate
## Показатель отказов (Bounce Rate) здесь составляет ~75% и выше.

# Конверсия сессии крайне низкая
## Поскольку для совершения покупки обычно нужно минимум 3–5 событий (например: view - view - cart - purchase), а 75% сессий обрываются на первом шаге, конверсия на уровне сессии будет мизерной. 
## Реальная конверсия считается на уровне пользователя или воронки за период, а не отдельного визита.
```

```python
# Основные наблюдения:

## Масштаб диктует view
## Зеленая линия (view) задает диапазон оси Y (до 140 000+), из-за чего синяя (cart) и оранжевая (purchase) линии визуально «сплющены» у нижней границы. Это прямое отражение общей статистики: ~95% всех действий — просмотры.

# Абсолютная синхронность кривых
## Все три типа событий двигаются идентично. Форма графика не меняется: утренний подъем, дневное плато, резкий взлет к 15:00–16:00 и обрыв после 18:00. Линии не пересекаются и не расходятся веером.

# Единый пик для всей воронки
## Максимум для всех действий приходится на 16:00 UTC. В этот час фиксируется ~145 тыс. просмотров, ~4.5 тыс. добавлений в корзину и ~2.5 тыс. покупок. Спад начинается одновременно для всех этапов.

# Конверсия стабильна в течение суток
## Тот факт, что кривые повторяют друг друга без существенных расхождений, означает: доля пользователей, доходящих до корзины и покупки, не зависит от времени дня. Вечером покупают больше не потому, что аудитория решительнее, а просто потому, что на сайте в 3–4 раза больше людей. 
## Воронка работает с одинаковой эффективностью утром, днем и вечером.

# Окончательное подтверждение таймзоны
## Синхронный пик в 16:00 UTC (~18:00–19:00 по Москве/Киеву/Минску) железобетонно указывает на географическое ядро аудитории.
## Это время возвращения домой, когда заканчиваются рабочие задачи и начинается личный шопинг.
```
